```python
import pandas as pd
```

```python
df = pd.read_csv("ExampleData.csv")
```

```python
df
```

```python
# Boş verilerin doldurmanın yolları
# Ortalama, orta değer, en yakın komşularına bak, en çok tekrar eden, regression

```

```python
from sklearn.impute import SimpleImputer
```

```python
imp=SimpleImputer(strategy= 'mean') # ortalama değer
```

```python
imp.fit_transform(df) # verileri doldurduk, fakat anlamsız görünüyor
```

```python
yenidata= imp.fit_transform(df) # anlamsız verileri yeni dataya attık 
```

```python
pd.DataFrame(yenidata) # verileri anlamlı görünür hale getirdik
```

```python
# en yakın komşu ile verileri doldurma
```

```python
from sklearn.impute import KNNImputer
```

```python
knn= KNNImputer()
```

```python
dolu = knn.fit_transform(df)
```

```python
pd.DataFrame(dolu) # dataları en yakın komşu ile doldurduk
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
imp=IterativeImputer(estimator=lr)
imputed_data=imp.fit_transform(df)
```

```python
pd.DataFrame(imputed_data)
```
