# Normalization Data and Principle Component Analysis

```python
import pandas as pd
```

```python
df = pd.read_csv('PastHires.csv')
```

```python
df
```

```python
# Employed tablosunda ki Y=1 N=0 , Level of Education tablosundaki verileri de 1-2-3 olarak değiştireceğiz
```

```python
df['Employed?']=df['Employed?'].map({'Y' : 1 , 'N' : 0})
```

```python
df
```

```python
# tek satırda değiştirmek istediğimiz verileri değiştirme komutu
```

```python
df = df.replace({'Y' : 1 , 'N' : 0, 'BS' : 1 , 'MS' : 2 , 'PhD' : 3})
```

```python
df
```

```python
# datayı normalize edeceğiz (belirli bir değer aralığına indirgiyoruz) her veriden min değeri çıkart / max - min değer
```
